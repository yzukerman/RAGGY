# RAG with Weaviate and your data

In this step, we will:
1. Get the user's query
2. Query the database
3. Send the query results with the query and a prompt to the LLM

## Get the user query

```python
query = input("Enter your query:")
```

```python
print("Your query was: " + query)
```

## Prep the LLM prompt
The LLM prompt will combine the query and the context we retrieve from the database. Below is a simple prompt template:

```python
PROMPT_TEMPLATE = """
Answer the question based only on the following context:
{context}

---
Answer the question based on the above context: {query}
"""
```

## Search the database

```python
import weaviate
import os
import weaviate.classes.query as wq

context = ""

headers = {
    "X-OpenAI-Api-Key": os.environ.get("OPENAI_API_KEY")
    #os.getenv("OPENAI_APIKEY")
}  # Replace with your OpenAI API key

try:    
    # connect to database
    client = weaviate.connect_to_local(headers=headers)
    movies = client.collections.get("ADI_DOCS")

    # Perform query
    response = movies.query.near_text(
        query=query, 
        limit=10, # maximum number of results
        return_metadata=wq.MetadataQuery(distance=True)
    )

    for o in response.objects:
        print(o.properties['chunk_content'])
        print("\n")
        context = context + o.properties['chunk_content'] + '\n---\n'

finally:
    client.close()
```

```python
PROMPT_TEMPLATE = f"""
Answer the question based only on the following context:
{context}

---
Answer the question based on the above context: {query}
"""
```

## Call the LLM with the prompt

```python
print (PROMPT_TEMPLATE)
```

## Call the LLM with the prompt

```python
from openai import OpenAI
import os
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))
MODEL="gpt-3.5-turbo-instruct"

completion = client.completions.create(
  model = MODEL,
    prompt = PROMPT_TEMPLATE,
    max_tokens = 150,
    temperature = 0.7,
)

print("Assistant: " + completion.choices[0].text.strip())
```
